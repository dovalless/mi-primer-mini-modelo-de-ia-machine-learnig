```python
# ==========================================
# MINI CHATGPT CON RED NEURONAL (TENSORFLOW)
# + VISUALIZACIÃ“N AVANZADA DE NEURONAS
# ==========================================

!pip install tensorflow scikit-learn nltk matplotlib seaborn

import nltk
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.utils import to_categorical

nltk.download("punkt")

# ==========================================
# DATA DE ENTRENAMIENTO
# ==========================================

texts = [
    "hola",
    "buenos dias",
    "como estas",
    "quien eres",
    "que es inteligencia artificial",
    "que es programacion",
    "adios"
]

intents = [
    "saludo",
    "saludo",
    "estado",
    "identidad",
    "ia",
    "programacion",
    "despedida"
]

responses = {
    "saludo": "Hola ðŸ‘‹ Â¿en quÃ© puedo ayudarte?",
    "estado": "Estoy funcionando correctamente ðŸ¤–",
    "identidad": "Soy un mini modelo de inteligencia artificial",
    "ia": "La IA es la capacidad de una mÃ¡quina para aprender patrones",
    "programacion": "Programar es darle instrucciones a una computadora",
    "despedida": "Hasta luego ðŸ‘‹"
}

# ==========================================
# PREPROCESAMIENTO
# ==========================================

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts).toarray()

encoder = LabelEncoder()
y = encoder.fit_transform(intents)
y = to_categorical(y)

vocab = vectorizer.get_feature_names_out()

# ==========================================
# RED NEURONAL
# ==========================================

input_layer = Input(shape=(X.shape[1],))
hidden1 = Dense(16, activation="relu", name="Capa_Oculta_1")(input_layer)
hidden2 = Dense(16, activation="relu", name="Capa_Oculta_2")(hidden1)
output_layer = Dense(y.shape[1], activation="softmax", name="Salida")(hidden2)

model = Model(inputs=input_layer, outputs=output_layer)

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.fit(X, y, epochs=300, verbose=0)
print("âœ… Modelo entrenado correctamente")

# ==========================================
# MODELO PARA ACTIVACIONES
# ==========================================

activation_model = Model(
    inputs=model.input,
    outputs=[layer.output for layer in model.layers]
)

# ==========================================
# FUNCIÃ“N DE CHAT INTERACTIVO
# ==========================================

def chat():
    print("\nðŸ¤– Mini ChatGPT Neural")
    print("Escribe 'salir' para terminar\n")

    while True:
        text = input("TÃº: ").lower()

        if text == "salir":
            print("ðŸ¤– Hasta luego")
            break

        vector = vectorizer.transform([text]).toarray()

        # PredicciÃ³n
        prediction = model.predict(vector, verbose=0)
        intent = encoder.inverse_transform([np.argmax(prediction)])[0]
        print("ðŸ¤–:", responses[intent])

        # ==========================================
        # VISUALIZACIÃ“N DE ACTIVACIONES
        # ==========================================

        activations = activation_model.predict(vector, verbose=0)

        plt.figure(figsize=(14,4))

        for i, activation in enumerate(activations[1:-1]):
            plt.subplot(1, len(activations)-2, i+1)
            sns.heatmap(
                activation,
                cmap="viridis",
                cbar=True,
                xticklabels=[f"N{i+1}" for i in range(activation.shape[1])],
                yticklabels=False
            )
            plt.title(f"ActivaciÃ³n {model.layers[i+1].name}")

        plt.tight_layout()
        plt.show()

        # ==========================================
        # EXPLICACIÃ“N DEL INPUT
        # ==========================================

        print("\nðŸ§  Palabras activas en el input:")
        for i, val in enumerate(vector[0]):
            if val > 0:
                print(f"  â€¢ '{vocab[i]}' â†’ activÃ³ la red")

        print("\n" + "-"*50 + "\n")

# ==========================================
# INICIAR CHAT
# ==========================================

chat()
```
